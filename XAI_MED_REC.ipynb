# **Clean and Change the Data**

```python
import re
def extract_side_effects(input_text):
    if not isinstance(input_text, str):
        return []
    regex_pattern = r'(?<=[a-z )])(?=[A-Z])'
    effects_list = re.split(regex_pattern, input_text)
    effects_list = [effect.strip() for effect in effects_list if effect.strip()]
    return effects_list
```

```python
dataset['Side_effects'] = dataset['Side_effects'].apply(extract_side_effects)
```

```python
dataset.head(10)
```

```python
def find_availability(input_text):
    if isinstance(input_text, str):
        regex_pattern = r"(available|sold out)"
        found_match = re.search(regex_pattern, input_text, re.I)
        if found_match:
            return found_match.group(0).lower()
        else:
            return "Not found"
    else:
        return "Not found"
```

```python
dataset['Status'] = dataset['Status'].apply(find_availability)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=20000, stop_words='english')
```

```python
selected_columns = dataset[['Uses','Composition', 'Side_effects','Alcohol','Pregnancy','Breast feeding','Driving','Kidney','Liver','Unit Type']]
text_data = selected_columns.apply(lambda row: ' '.join(row.astype(str)), axis=1)
```

```python
vector = tfidf.fit_transform(text_data).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity.shape
```

```python
def recommendation_system_cosin(medicine, dataset, similarity, price_limit=None):
    # fuzzy matching to find the closest match to the input medicine
    fuzzy_matches = process.extractOne(medicine, dataset['Name'].values)
    closest_match = fuzzy_matches[0]
    
    # Finding the index of the closest match in the dataset
    if closest_match in dataset['Name'].values:
        med_index = dataset[dataset['Name'] == closest_match].index[0]
        
        # Retrieve the similarity scores for the closest match
        distances = similarity[med_index]
        
        # Get indices of the top 5 similar medicines (excluding the input medicine)
        med_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        
        # Collect the top 5 similar medicines with required information
        recommended_medicines = []
        seen_medicines = set()  # To keep track of medicines we've already added
        
        for i in med_list:
            med_index = i[0]
            similarity_score = round(i[1], 2)
            med_info = dataset.iloc[med_index]
            
            # Check if the medicine is already in the recommended list (to avoid duplicates)
            if med_info['Name'] not in seen_medicines:
                if price_limit is None or med_info['Price'] <= price_limit:
                    recommended_medicines.append({
                        "Name": med_info['Name'],
                        "Side_effects": med_info['Side_effects'],
                        "Alcohol": med_info['Alcohol'],
                        "Pregnancy": med_info['Pregnancy'],
                        "Breast_feeding": med_info['Breast feeding'],
                        "Driving": med_info['Driving'],
                        "Kidney": med_info['Kidney'],
                        "Liver": med_info['Liver'],
                        "Price": med_info['Price'],
                        "Similarity_Score": similarity_score,
                        "Uses": med_info['Uses']  # Add "Uses" field
                    })
                    seen_medicines.add(med_info['Name'])  # Add to seen set

        return recommended_medicines
```

```python
# Get user input for the medicine name and price limit
medicine_name = input("Enter the tablet name: ")
try:
    price_limit = float(input("Enter your price limit (leave blank if no limit): ")) if input("Do you want to set a price limit? (y/n): ").lower() == 'y' else None
except ValueError:
    print("Invalid input for price limit. Continuing without a price limit.")
    price_limit = None

# Call the recommendation function and print results
recommendations = recommendation_system_cosin(medicine_name, dataset, similarity, price_limit)

# Display recommendations
if recommendations:
    print(f"\nTop Recommendations for '{medicine_name}':")
    for rec in recommendations:
        print(f"{rec['Name']} - Similarity Score: {rec['Similarity_Score']}")
        print(f"   Side Effects: {rec['Side_effects']}")
        print(f"   Alcohol: {rec['Alcohol']}, Pregnancy: {rec['Pregnancy']}, Breast Feeding: {rec['Breast_feeding']}, Driving: {rec['Driving']}, Kidney: {rec['Kidney']}, Liver: {rec['Liver']}")
        print(f"   Price: {rec['Price']} ")
        print(f"   Uses: {rec['Uses']}\n")  # Print the Uses information
else:
    print("No recommendations found within the specified criteria.")
```
